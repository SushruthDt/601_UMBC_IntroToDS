```python
data1 = pd.read_excel("C:\\Users\\lenovo\\Documents\\GitHub\\601_UMBC_IntroToDS\\Week 9\\imputation_homework_12.xls") #importing data
data1.head()
```

```python
data1.shape #Number of Rows and columns
```

```python
data1.isnull().sum() #Number of NA values
```

```python
data1.plot(kind = 'hist',alpha=0.4) #Plot the distribution of data points using a histogram
fig = plt.gcf()
fig.set_size_inches(10,5)
plt.show()
```

```python
pd.plotting.lag_plot(data1['value'],lag=1); #lag plot (a lag plot shows t versus t+1)
fig = plt.gcf()
fig.set_size_inches(10,5)
plt.show()
```

Observation - The data points are clustered along positive trending belt from bottom left to top right i.e  upward diagonal path .
I see a order of data points

```python
plt.scatter(data1['time'],data1['value']);
plt.xlabel('time')
plt.ylabel('value')
fig = plt.gcf()
fig.set_size_inches(15,8)
plt.show()
```

Observation - The data points are clustered in orderely manner of roughly sinusoidal shape.I see a order of data points. Hence I would try to then interpolate the missing values

```python
data_intra = data1.copy() #copying from original data
```

```python
data_intra['value']= data_intra['value'].interpolate()
```

```python
data_intra.isnull().sum() #Number of NA values
```

```python
data_intra.plot(kind = 'hist',alpha=0.4) #Plot the distribution of data points using a histogram
fig = plt.gcf()
fig.set_size_inches(10,5)
plt.show()
```
